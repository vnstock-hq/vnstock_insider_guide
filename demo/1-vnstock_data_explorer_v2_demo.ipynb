## 6. Giá hàng hoá

```python
from vnstock_data import CommodityPrice

# Khởi tạo đối tượng CommodityPrice với ngày mặc định
commodity = CommodityPrice(start="2022-01-03", end="2024-12-31")

# Lấy dữ liệu giá vàng
gold_vn = commodity.gold_vn()  # Giá vàng Việt Nam (mua vào)
print("Gold VN:\n", gold_vn)

gold_global = commodity.gold_global()  # Giá vàng thế giới
print("Gold Global:\n", gold_global)

# Lấy dữ liệu giá xăng
gas_vn = commodity.gas_vn()  # Giá xăng RON92
print("Gas VN:\n", gas_vn)

oil_crude = commodity.oil_crude()  # Giá dầu thô
print("Oil Crude:\n", oil_crude)

# Lấy dữ liệu khí thiên nhiên
gas_natural = commodity.gas_natural()  # Giá khí thiên nhiên
print("Gas Natural:\n", gas_natural)

# Lấy dữ liệu giá than
coke = commodity.coke()  # Giá than cốc
print("Coke:\n", coke)

# Lấy dữ liệu giá thép
steel_d10 = commodity.steel_d10()  # Giá thép D10
print("Steel D10:\n", steel_d10)

steel_hrc = commodity.steel_hrc()  # Giá thép HRC
print("Steel HRC:\n", steel_hrc)

# Lấy dữ liệu giá quặng sắt
iron_ore = commodity.iron_ore()  # Giá quặng sắt
print("Iron Ore:\n", iron_ore)

# Lấy dữ liệu giá phân bón
fertilizer_ure = commodity.fertilizer_ure()  # Giá phân ure
print("Fertilizer URE:\n", fertilizer_ure)

# Lấy dữ liệu giá nông sản
soybean = commodity.soybean()  # Giá đậu tương
print("Soybean:\n", soybean)

corn = commodity.corn()  # Giá ngô (bắp)
print("Corn:\n", corn)

sugar = commodity.sugar()  # Giá đường
print("Sugar:\n", sugar)

# Lấy dữ liệu giá heo hơi
pork_north_vn = commodity.pork_north_vn()  # Giá heo hơi miền Bắc Việt Nam
print("Pork North VN:\n", pork_north_vn)

pork_china = commodity.pork_china()  # Giá heo hơi Trung Quốc
print("Pork China:\n", pork_china)
```
